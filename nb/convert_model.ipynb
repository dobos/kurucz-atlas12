# Converting models

Read in `atlas9` grid model files, split into separate models by T_eff and prefix with parameters necessary for `SYNTHE`

```python
import os
```

```python
synthe_header = \
"""SURFACE INTENSI 17 1.,.9,.8,.7,.6,.5,.4,.3,.25,.2,.15,.125,.1,.075,.05,.025,.01
ITERATIONS 1 PRINT 2 PUNCH 2
CORRECTION OFF
PRESSURE OFF
MOLECULES ON
READ MOLECULES
"""
```

```python
infilename = '/export/data/dobos/data/pfsspec/source/stellar/grid/atlas9/gridm10odfnew/am10k2odfnew.dat'
outdir = '/export/data/dobos/data/pfsspec/source/stellar/grid/synthe'
modname = os.path.basename(infilename)
modname = os.path.splitext(modname)[0]
modname
```

```python
with open(infilename, 'r') as f:
    buffer = ""
    for line in f:
        buffer += line
        if line[:4] == 'TEFF':
            t_eff = float(line[4:13])
            log_g = float(line[21:29])
        #if line[:5] == 'TITLE':
        #    title = line[6:]
        if line[:5] == 'BEGIN':
            if (t_eff <= 5000):
                outfilename = modname + str(int(t_eff)) + 'g'
                if log_g < 0:
                    outfilename += 'm'
                else:
                    outfilename += 'p'
                outfilename += '{:02d}'.format(int(abs(log_g * 10)))
                outfilename += '.mod'
                outfilename = os.path.join(outdir, outfilename)
                with open(outfilename, 'w') as w:
                    w.write(synthe_header)
                    w.write(buffer)
            buffer = ""
            print(title)
```
